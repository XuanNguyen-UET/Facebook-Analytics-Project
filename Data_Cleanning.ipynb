#### 2)Comment_full to new DataFrame: Comment_frame
##### Convert comment_full cols into a Dataframe with 4 fields: 
- comment_id: 
- commenter_id:
- comment_text: 
- post_id: 

```python
def get_comments(text):
    pattern = r"\{[^{}]*\}"
    try:
        matches = re.findall(pattern, text)
        return [eval(match) for match in matches]
    except Exception as e:
        print(f"Err: {e}")
        return []

def create_comments_frame(df):
    df_result = {
        "comment_id": [],
        "commenter_id": [],
        "commenter_name": [],
        "comment_text": [],
        "post_id": [],
    }

    new_df = df[~df["comments_full"].isna()]

    for index, row in new_df.iterrows():
        cmt = row["comments_full"]
        for comment in get_comments(cmt):
            try:
                df_result["comment_id"].append(comment.get("comment_id"))
                df_result["comment_text"].append(comment.get("comment_text"))
                df_result["commenter_name"].append(comment.get("commenter_name"))
                df_result["commenter_id"].append(comment.get("commenter_id"))
                df_result["post_id"].append(row.post_id)
            except:
                continue

    return pd.DataFrame(df_result)
```

```python
Comment_frame = create_comments_frame(clean_df)
Comment_frame
```

```python
# drop rows that have no comment_id
Comment_frame = Comment_frame.dropna(subset=['comment_id'])
Comment_frame
```

#### 3)Create Dataframe for reactions from column "reactions"

```python
dfs = []
for index, row in clean_df.iterrows():
    if row['reactions'] is not np.nan:
        # Convert the 'reactions' column into a DataFrame
        df = pd.DataFrame(eval(row['reactions']), index=[0])
        df['post_id'] = row['post_id']
        dfs.append(df)

# Concat
Reactions_Frame = pd.concat(dfs, ignore_index=True)

# Replace NaN values with 0
Reactions_Frame = Reactions_Frame.fillna(0)

# Convert data type to int
Reactions_Frame = Reactions_Frame.astype(int)

# Addthe reactions of each post
Reactions_Frame['total'] = Reactions_Frame.drop('post_id', axis=1).sum(axis=1)
```

#### remove blank line (\n) , emojis in text

```python
def clean_text(text):
    text = text.replace("\n", " ")
    text = text.replace("\t", " ")

    # Remove emojis
    emoj = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642" 
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
                      "]+", re.UNICODE)
    text = re.sub(emoj, '', text)

    return text
```

#### convert time to datetime 

```python
# convert time to datetime
clean_df["time"] = pd.to_datetime(clean_df["time"])
```

```python
# clean the text
clean_df["text"] = clean_df["text"].astype(str).apply(clean_text)
Comment_frame["comment_text"] = Comment_frame["comment_text"].astype(str).apply(clean_text)
```
